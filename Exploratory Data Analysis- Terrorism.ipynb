```python
plt.figure(figsize=(20,10))
sns.countplot(y=mydata1['attacktype1_txt'], hue=mydata1['targtype1_txt'], order=mydata['attacktype1_txt'].value_counts().index, palette="Set3");
```

The above plot explains about the attack type and target of terrorists. We can see Private Citizens & property and military has been attacked majority of times. The most popular weapon used by terrorists are bombing/explosion and armed assault.

```python
plt.figure(figsize=(10,25))
plt.pie(mydata1['weaptype1_txt'].value_counts(), autopct="%1.1f%%");
#plt.tight_layout()
plt.legend(labels=mydata1['weaptype1_txt'].value_counts().index,loc='upper left', fontsize='small');
```

The pie plot talks about distribution of weapon type data. We can see explosives are used in more amount than others for terrorist attacks. 

```python
ct1=pd.crosstab(mydata['attacktype1_txt'], mydata['region_txt'])
ct1
```

```python
fig,ax=plt.subplots(figsize=(10,8));
ct1.plot.bar(stacked=True, ax=ax);
```

The bar chart shows South Asia and Middele East & North Africa has been attacked many times by using Armed Assault and Bombing/Explosions. Terrorist doesn't usually prefer Hostage taking, Unarmed assault and hijacking.

```python
ax,fig=plt.subplots(figsize=(15,30))
sns.countplot(y=mydata['country_txt'], order=mydata['country_txt'].value_counts().index);
```

The above chart gives a count of terrorist attack across different countries. Here Iraq stands first following Pakistan and Afghanistan. Terrorist attacks are witnessed much in these countries. 

```python
sns.distplot(mydata1['iyear'], kde=True, hist=False);
```

The above plot shows increase in terrorist attacks. There's a drastic drop of number of terrorist attacks from 2015.
